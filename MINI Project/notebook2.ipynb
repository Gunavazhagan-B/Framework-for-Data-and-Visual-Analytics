```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pickle
```

```python
ordinal=OrdinalEncoder()
label=LabelEncoder()
scaler=StandardScaler()
```

```python
ordinal_cols = ['week_of_outbreak']
label_cols = ['district']
ohe_cols = ['state_ut']

# --- Step 2: Handle LabelEncoder separately (since it doesn't fit directly in ColumnTransformer) ---
label_encoder = LabelEncoder()
df['district_encoded'] = label_encoder.fit_transform(df['district'])

# Replace the original 'district' column
df = df.drop(columns=['district'])
df['district'] = df['district_encoded']
df = df.drop(columns=['district_encoded'])

# --- Step 3: Define sub-pipelines for the remaining encoders ---
ordinal_pipeline = Pipeline(steps=[
    ('ordinal', OrdinalEncoder())
])

ohe_pipeline = Pipeline(steps=[
    ('ohe', OneHotEncoder(sparse_output=False, drop='first'))
])

# --- Step 4: Combine everything in a ColumnTransformer ---
preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', ordinal_pipeline, ordinal_cols),
        ('ohe', ohe_pipeline, ohe_cols)
    ],
    remainder='passthrough'  # Keeps all other columns as they are
)

# --- Step 5: Fit the preprocessor on the data ---
preprocessor.fit(df)

# --- Step 6: Save both preprocessor and label encoder ---
with open("preprocessing_pipeline.pkl", "wb") as f:
    pickle.dump(preprocessor, f)

with open("label_encoder_district.pkl", "wb") as f:
    pickle.dump(label_encoder, f)

print("✅ Preprocessing pipeline and label encoder saved successfully!")
```
