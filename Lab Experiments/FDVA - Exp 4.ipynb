```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Datasets/StudentsPerformance.csv')

# Display first few rows
print("✅ First 5 Records:")
display(df.head())

# Display first 3 and last 5 records
print("\nFirst 3 Records:")
display(df.head(3))

print("\nLast 5 Records:")
display(df.tail())

# Display shape of dataset
print("\nDataset Shape:", df.shape)

# Display column names
print("\nColumn Names:", df.columns.tolist())

# Display data types
print("\nData Types:")
print(df.dtypes)

# Display detailed info
print("\nDataset Info:")
df.info()

print("\n---- Filtering and Subsetting ----")

# Students with math score > 70
print("\nStudents with math score > 70:\n")
display(df[df["math score"] > 70])

# Female students only
print("\nFemale Students:\n")
display(df[df["gender"] == "female"])

# Select only specific columns
print("\nSubset with 'gender' and 'math score':\n")
display(df[["gender", "math score"]])

print("\n---- Descriptive Statistics ----")

# Extract the math score column
math_scores = df["math score"]

# Compute statistical measures
mean = math_scores.mean()
median = math_scores.median()
mode = math_scores.mode()[0]   # mode() returns a Series
_range = math_scores.max() - math_scores.min()
variance = math_scores.var()
std_dev = math_scores.std()

# Print statistics
print(f"\nMean (Math Score): {mean:.2f}")
print(f"Median (Math Score): {median}")
print(f"Mode (Math Score): {mode}")
print(f"Range (Math Score): {_range}")
print(f"Variance (Math Score): {variance:.2f}")
print(f"Standard Deviation (Math Score): {std_dev:.2f}")

avg_scores = {
    "Math": df["math score"].mean(),
    "Reading": df["reading score"].mean(),
    "Writing": df["writing score"].mean()
}

plt.figure(figsize=(6, 4))
plt.bar(avg_scores.keys(), avg_scores.values(), color=['skyblue', 'lightgreen', 'salmon'])
plt.title("Average Scores per Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(df["math score"], bins=5, edgecolor="black", color='lightblue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(4, 4))
plt.boxplot(df["math score"], patch_artist=True, boxprops=dict(facecolor="lightyellow"))
plt.title("Boxplot of Math Scores")
plt.ylabel("Math Score")
plt.show()

plt.figure(figsize=(7, 4))
plt.hist(df["math score"], bins=5, edgecolor="black", alpha=0.6, color='lightgray')

# Add vertical lines for mean, median, and mode
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='green', linestyle='-.', linewidth=2, label=f"Median: {median}")
plt.axvline(mode, color='blue', linestyle=':', linewidth=2, label=f"Mode: {mode}")

plt.title("Math Score Distribution with Mean, Median, and Mode")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
